X
X минут после полуночи, однако для этого необходимо указать время сигнала в формате часы, минуты. Помогите Коле определить, на какое время завести будильник.

Часы и минуты в выводе программы должны располагаться на разных строках (см. пример работы программы)

Помните, что для считывания данных нужно вызывать функцию input без аргументов!

```python
xy = input()
print(int(xy)//60)
print(int(xy)%60)
```

X
X минут. В отличие от Коли, Катя ложится спать после полуночи в 
H
H часов и 
M
M минут. Помогите Кате определить, на какое время ей поставить будильник, чтобы он прозвенел ровно через 
X
X минут после того, как она ляжет спать.

На стандартный ввод, каждое в своей строке, подаются значения 
X
X, 
H
H и 
M
M. Гарантируется, что Катя должна проснуться в тот же день, что и заснуть. Программа должна выводить время, на которое нужно поставить будильник: в первой строке часы, во второй — минуты.

```python
x = int(input())
n = int(input())
m = int(input())

m = m+n*60+x

print(m//60)
print(m%60)
```
